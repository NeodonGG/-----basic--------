```python
import pandas as pd
import folium
from folium.plugins import FastMarkerCluster

# (이전에 df = pd.read_csv('listings.csv') 코드로 데이터를 불러왔다고 가정할게!)

Long = -73.80
Lat = 40.80

# 위도와 경도 데이터를 리스트 형태로 묶어주기
locations = list(zip(df.latitude, df.longitude))

# 중심 좌표(location)와 초기 확대 비율(zoom_start)을 설정하여 기본 지도 생성
map1 = folium.Map(location=[Lat, Long], zoom_start=9)

# 대량의 마커를 빠르게 렌더링하고 클러스터링(그룹화)해 주는 FastMarkerCluster 추가
FastMarkerCluster(data=locations).add_to(map1)

# 지도 출력
map1
```
